```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
all_data = pd.read_csv('../data/all_data.csv')
```

```python
all_data.head(3)
```

```python
# 'fuel' の欠損値を 'gas' で埋める
all_data['fuel'].fillna('gas', inplace=True)

all_data['title_status'].fillna('clean', inplace=True)

all_data['type'].fillna('sedan', inplace=True)

all_data['state'].fillna('ca', inplace=True)
```

```python
# LabelEncoderのインスタンスを作成
label_encoder = LabelEncoder()

# カテゴリ変数のリスト
columns_to_encode = ['region', 'manufacturer', 'condition', 'cylinders', 'fuel', 'title_status', 'transmission', 'drive', 'size', 'type', 'paint_color', 'state']

# 各カテゴリ変数をラベルエンコーディング
label_encoders = {}  # 各カテゴリ変数のエンコーダを保存するための辞書
for column in columns_to_encode:
    le = LabelEncoder()
    all_data[column] = le.fit_transform(all_data[column].astype(str))
    label_encoders[column] = le
```
